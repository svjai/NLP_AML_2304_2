```python
import pandas as pd


# Read the text file line by line
with open('VideoCommentsThreatCorpus.txt', 'r',encoding='utf-8') as file:
    lines = file.readlines()
    label=[]
#     print(lines)
    for line in lines:
        for j in line.split('\n'):
            if len(j.split('\t'))!=1:
                label.append(j.split('\t'))
            else:
                print(j) #for time being removing the metadata (details of the video and date)
```

```python
df=pd.DataFrame(label) #creating a dataframe from the created list
```

```python
df=df.iloc[:,:2] #filtering only the requried column
```

```python
df['len_text']=df[1].str.len() #adding length of the each comments as seprate column
```
